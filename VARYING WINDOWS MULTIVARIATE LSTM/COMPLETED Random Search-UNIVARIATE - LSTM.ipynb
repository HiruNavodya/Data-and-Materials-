```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
import tensorflow as tf
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.optimizers import Adam
from scipy.stats import loguniform

# Set seeds
random.seed(42)
np.random.seed(42)
tf.random.set_seed(42)

# Load data
sheet_id = "1j_Euo80PrGckVDVr2hTG9zZebxJD0TSC"
sheet_name = "Sheet1"
csv_url = f"https://docs.google.com/spreadsheets/d/{sheet_id}/gviz/tq?tqx=out:csv&sheet={sheet_name}"
df = pd.read_csv(csv_url)
df = df.set_index('Date')
df = df.drop(columns=["YEAR", "MO", "DY"])

results = []

def sample_parameters():
    """Sample parameters from distributions"""
    return {
        'n_units': random.randint(32, 128),          # Uniform integer between 32 and 128
        'dropout_rate': random.uniform(0.1, 0.4),    # Uniform float between 0.1 and 0.4
        'learning_rate': loguniform.rvs(1e-4, 1e-2),  # Log-uniform between 1e-4 and 1e-2
        'batch_size': random.choice([16, 32, 64, 128]), # Discrete choices
        'n_epochs': random.randint(20, 50)           # Uniform integer between 20 and 50
    }

# Number of random combinations to try
n_iter = 20

for K in range(1, 15):
    # Create lag features
    lagged_df = df.copy()
    for lag in range(1, K + 1):
        lagged_df[f'lag_{lag}'] = lagged_df['WS10M'].shift(lag)
    lagged_df.dropna(inplace=True)

    # Split into train/test
    train_size = int(len(lagged_df) * 0.7)
    train = lagged_df.iloc[:train_size]
    test = lagged_df.iloc[train_size:]

    X_train = train.drop(columns=['WS10M'])
    y_train = train['WS10M']
    X_test = test.drop(columns=['WS10M'])
    y_test = test['WS10M']

    # Scale
    scaler_X = MinMaxScaler()
    scaler_y = MinMaxScaler()
    X_train_scaled = scaler_X.fit_transform(X_train)
    X_test_scaled = scaler_X.transform(X_test)
    y_train_scaled = scaler_y.fit_transform(y_train.values.reshape(-1, 1)).flatten()
    y_test_scaled = scaler_y.transform(y_test.values.reshape(-1, 1)).flatten()

    # Reshape for LSTM
    X_train_scaled = X_train_scaled.reshape((X_train_scaled.shape[0], X_train_scaled.shape[1], 1))
    X_test_scaled = X_test_scaled.reshape((X_test_scaled.shape[0], X_test_scaled.shape[1], 1))

    best_score = float('inf')
    best_params = None

    # Random search
    for _ in range(n_iter):
        params = sample_parameters()

        model = Sequential([
            LSTM(params['n_units'], activation="tanh", input_shape=(K, 1)),
            Dropout(params['dropout_rate']),
            Dense(1)
        ])
        optimizer = Adam(learning_rate=params['learning_rate'])
        model.compile(optimizer=optimizer, loss="mean_squared_error")

        history = model.fit(X_train_scaled, y_train_scaled,
                         epochs=params['n_epochs'],
                         batch_size=params['batch_size'],
                         validation_split=0.2,
                         verbose=0,
                         callbacks=[tf.keras.callbacks.EarlyStopping(
                             monitor='val_loss',
                             patience=3,
                             restore_best_weights=True)])

        pred = model.predict(X_test_scaled).flatten()
        current_score = mean_squared_error(y_test_scaled, pred)

        if current_score < best_score:
            best_score = current_score
            best_params = params
            print(f"New best params for K={K}: {best_params} with MSE: {best_score:.4f}")

    # Train final model with best parameters (full training set)
    final_model = Sequential([
        LSTM(best_params["n_units"], activation="tanh", input_shape=(K, 1)),
        Dropout(best_params["dropout_rate"]),
        Dense(1)
    ])
    optimizer = Adam(learning_rate=best_params["learning_rate"])
    final_model.compile(optimizer=optimizer, loss="mean_squared_error")

    history = final_model.fit(X_train_scaled, y_train_scaled,
                  epochs=100,
                  batch_size=best_params["batch_size"],
                  validation_split=0.2,
                  verbose=0,
                  callbacks=[tf.keras.callbacks.EarlyStopping(
                      monitor='val_loss',
                      patience=5,
                      restore_best_weights=True)])

    y_pred_test_scaled = final_model.predict(X_test_scaled).flatten()
    y_pred_test = scaler_y.inverse_transform(y_pred_test_scaled.reshape(-1, 1)).flatten()

    mse = mean_squared_error(y_test, y_pred_test)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_test, y_pred_test)
    mape = np.mean(np.abs((y_test - y_pred_test) / y_test)) * 100
    r2 = r2_score(y_test, y_pred_test)

    results.append({
        "Lag": K,
        "MSE": mse,
        "RMSE": rmse,
        "MAE": mae,
        "MAPE": mape,
        "R2": r2,
        "Best_Params": best_params,
        "Actual_Epochs_Used": len(history.history['loss'])
    })

    # Plot predictions
    plt.figure(figsize=(10, 4))
    plt.plot(y_test.values, label='Actual')
    plt.plot(y_pred_test, label='Predicted')
    plt.title(f'LSTM Prediction (Lag={K})')
    plt.xlabel('Time Step')
    plt.ylabel('WS10M')
    plt.legend()
    plt.grid()
    plt.tight_layout()
    plt.show()

results_df = pd.DataFrame(results)
results_df
```
